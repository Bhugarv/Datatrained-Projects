## Appying Machine Learning

```python
from sklearn.model_selection import train_test_split
```

```python
x=dfn.drop('Salary',axis=1)
x
```

```python
y=dfn['Salary']
y
```

### Liner Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
from sklearn.metrics import r2_score
```

```python
for i in range(0,1000,1):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.25)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(r2_score(y_train,pred_train)*100,1)==round(r2_score(y_test,pred_test)*100,1):
        print("Optimal Random State is: ",i)
        print("Training R2 score is",r2_score(y_train,pred_train)*100)
        print("Testing R2 score is",r2_score(y_test,pred_test)*100)
else:
    print("Optimal state not found")
```

```python
max_scr=0
for i in range(0,10000):
    x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(r2_score(y_train,pred_train)*100,2) == round(r2_score(y_test,pred_test)*100,2):
        print( "At random state",  i,  "The model performs well")
        print("Training r2 score is ", r2_score(y_train,pred_train)*100)
        print("Testing r2 score is ", r2_score(y_test,pred_test)*100)
        print("---------------------------")
```
