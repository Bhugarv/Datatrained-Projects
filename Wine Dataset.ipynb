```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR= Q3-Q1
print(IQR)
```

```python
df_new1 = df[~((df< (Q1 - 1.5* IQR)) | (df> (Q3 +1.5 * IQR))).any(axis=1)]
print(df_new1.shape)
```

## Applying ML algos

```python
x=df.drop(['Class'],axis=1)
x.head()
```

```python
y=df['Class']
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
max_cv=0
for i in range(0,100,2):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
        print("At random State - ",i,"model is performing good")
        print("Traing Accuracy sore is - ",accuracy_score(y_train,pred_train)*100)
        print("Testing Accuracy sore is - ",accuracy_score(y_test,pred_test)*100)
    
```

```python
x_scaled=sc.fit_transform(x)
```

```python
print("At random State - 78 model is performing good")
print("Traing Accuracy sore is - ",accuracy_score(y_train,pred_train)*100)
print("Testing Accuracy sore is - ",accuracy_score(y_test,pred_test)*100)
```

```python
x_train_n,x_test_n,y_train_n,y_test_n=train_test_split(x_scaled,y,random_state=78,test_size=0.25)
```

## Using Decission Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train_n,y_train_n)
```

```python
dt.score(x_test,y_test)
```

```python
grid_param={'criterion': ['gini', 'entropy'],
    'max_depth' : range(2,32,1),
    'min_samples_leaf' : range(1,10,1),
    'min_samples_split': range(2,10,1),
    'splitter' : ['best', 'random']}
```

```python
grid_search= GridSearchCV(dt,param_grid=grid_param,cv=5)
```

```python
grid_search.fit(x_train_n,y_train_n)
```

```python
best_param = grid_search.best_params_
print(best_param)
```

```python
dtn= DecisionTreeClassifier(criterion='entropy',max_depth=9,min_samples_leaf=1,min_samples_split=5,splitter='random')
dtn.fit(x_train,y_train)
```
